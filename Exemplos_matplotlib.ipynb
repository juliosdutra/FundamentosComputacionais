# ðŸŽ¨ IntroduÃ§Ã£o Ã  VisualizaÃ§Ã£o de Dados com Matplotlib

A biblioteca **Matplotlib** Ã© uma das ferramentas mais populares em Python para criaÃ§Ã£o de grÃ¡ficos. Ela permite representar dados de forma visual, o que facilita a anÃ¡lise, a comunicaÃ§Ã£o e a compreensÃ£o de informaÃ§Ãµes complexas.

## ðŸ§­ Por que usar grÃ¡ficos?

- ðŸ“ˆ Compreender **tendÃªncias e padrÃµes**
- ðŸ” Investigar **relaÃ§Ãµes entre variÃ¡veis**
- ðŸ§ª Visualizar **resultados de experimentos ou simulaÃ§Ãµes**
- ðŸ§  Apoiar **decisÃµes tÃ©cnicas** e **comunicaÃ§Ã£o cientÃ­fica**

## ðŸ§© Anatomia de um grÃ¡fico

Um grÃ¡fico em Matplotlib geralmente envolve:

- **Figura (`Figure`)**: a Ã¡rea total de desenho (janela do grÃ¡fico)
- **Eixos (`Axis`)**: regiÃ£o onde os dados sÃ£o desenhados (com eixos x e y)
- **Curvas e marcadores (`Lines`)**: os dados propriamente ditos
- **RÃ³tulos (`Labels`)**: nomes dos eixos, tÃ­tulo e legendas
- **Grade (`Grid`)**: linhas de referÃªncia opcionais
- **Barra de cores (`Colorbar`)**: Ãºtil em mapas de calor

# **ðŸ“ˆ GrÃ¡fico de linhas - `plt.plot()`**

O `plt.plot()` Ã© usado para visualizar dados contÃ­nuos, como variaÃ§Ãµes ao longo do tempo ou resultados de simulaÃ§Ãµes.

Este exemplo mostra como sobrepor duas curvas simples, aplicando boas prÃ¡ticas de visualizaÃ§Ã£o: uso de `label`, `grid`, tÃ­tulo, rÃ³tulos de eixos e `tight_layout()` para melhor espaÃ§amento.



````python
# importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# criar variÃ¡vel x (independente)
var_x = np.array([0, 1, 2, 3])

# criar variÃ¡veis y (dependentes)
y1 = [0, 1, 2, 3]                      # funÃ§Ã£o linear: y = x
y2 = [x**2 + 1 for x in var_x]         # funÃ§Ã£o quadrÃ¡tica: y = xÂ² + 1

# criar a figura (tamanho opcional)
plt.figure(1, figsize=(6, 3))

# plotar os grÃ¡ficos com rÃ³tulos e estilos
plt.plot(var_x, y1, 'b-', label='Linear')        # linha azul contÃ­nua
plt.plot(var_x, y2, 'r--', label='QuadrÃ¡tica')   # linha vermelha tracejada

# customizaÃ§Ã£o dos eixos e tÃ­tulo
plt.xlabel('VariÃ¡vel independente')
plt.ylabel('Valor de resposta')
plt.title('Exemplo de GrÃ¡fico')

# exibir legenda e grade
plt.legend()
plt.grid(True)

# ajustar o layout e exibir
plt.tight_layout()
plt.show()
````

# **MÃºltiplos grÃ¡ficos - `plt.subplot(m,n,p)`**

O comando `plt.subplot(m, n, p)` divide a figura em uma grade de `m` linhas e `n` colunas, e ativa a posiÃ§Ã£o `p` para desenhar o grÃ¡fico. Isso Ã© Ãºtil para comparar diferentes visualizaÃ§Ãµes lado a lado.

````python
# Criar vetor de entrada
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)          # funÃ§Ã£o seno
y2 = np.cos(x)          # funÃ§Ã£o cosseno
y3 = np.sin(x) + np.cos(x)  # combinaÃ§Ã£o das duas

# Criar figura
plt.figure(figsize=(6, 6))

# GrÃ¡fico 1: seno
plt.subplot(3, 1, 1)  # 3 linhas, 1 coluna, 1Âª posiÃ§Ã£o
plt.plot(x, y1, 'b')
plt.title('FunÃ§Ã£o seno')
plt.ylabel('sin(x)')

# GrÃ¡fico 2: cosseno
plt.subplot(3, 1, 2)  # 3 linhas, 1 coluna, 2Âª posiÃ§Ã£o
plt.plot(x, y2, 'r')
plt.title('FunÃ§Ã£o cosseno')
plt.ylabel('cos(x)')

# GrÃ¡fico 3: soma
plt.subplot(3, 1, 3)  # 3 linhas, 1 coluna, 3Âª posiÃ§Ã£o
plt.plot(x, y3, 'g')
plt.title('Seno + Cosseno')
plt.xlabel('x')
plt.ylabel('sin(x) + cos(x)')

# Ajuste automÃ¡tico para evitar sobreposiÃ§Ã£o
plt.tight_layout()
plt.show()
````

# **GrÃ¡fico em 2D/3D**

## **Como usar `plt.contourf()` e `plot_surface`?**

> ðŸ’¡ ExplicaÃ§Ã£o RÃ¡pida:
* `contourf` â†’ visualizaÃ§Ã£o 2D com nÃ­veis de cor preenchidos.
* `plot_surface` â†’ representaÃ§Ã£o 3D da mesma funÃ§Ã£o.
* `cm.viridis_r` â†’ colormap invertido para melhor contraste.

**Exemplo simples**

````python
Z = np.random.uniform(0,1,(8,8))
plt.contourf(Z, cmap='viridis_r')
plt.colorbar()
plt.show()
````

**Exemplo com funÃ§Ã£o $Z=f(X,Y)$**

````python

from mpl_toolkits.mplot3d import Axes3D

# Criar a grade de pontos (x, y)
nx, ny = 100, 100
x = np.linspace(0, 0.909, nx)
y = np.linspace(0, 0.909, ny)
X, Y = np.meshgrid(x, y)

# FunÃ§Ã£o paraboloide deslocado
Z = (X - 0.2)**2 + (Y - 0.2)**2

# Criar figura com 2 subplots lado a lado
fig = plt.figure(figsize=(12, 5))

# Subplot 1: contorno (2D)
ax1 = fig.add_subplot(1, 2, 1)
cont = ax1.contourf(x, y, Z, cmap='viridis_r')  # mapa de cores
cbar1 = fig.colorbar(cont, ax=ax1)  # barra de cores
cbar1.set_label('z = f(x, y)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Mapa de Contorno')

# Subplot 2: superfÃ­cie (3D)
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax2.plot_surface(X, Y, Z, cmap='viridis_r')
cbar2 = fig.colorbar(surf, ax=ax2, shrink=0.5, aspect=10, pad=0.1)
cbar2.set_label('z = f(x, y)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z = f(x, y)')
ax2.set_title('Paraboloide')

# Ajuste de layout para evitar sobreposiÃ§Ã£o
plt.tight_layout()
plt.show()
````

## **ðŸŸ¦ Ãrea Sombreada - `plt.fill_between()`**


O `plt.fill_between()` preenche a Ã¡rea entre duas curvas ou entre uma curva e um valor fixo.

Ã‰ ideal para mostrar:
- Incertezas experimentais
- Intervalos de confianÃ§a
- Faixas de operaÃ§Ã£o segura

````python
x = np.linspace(0, 10, 100)
y_central = np.sin(x)
y_sup = y_central + 0.3
y_inf = y_central - 0.3

plt.plot(x, y_central, label='MÃ©dia', color='blue')
plt.fill_between(x, y_inf, y_sup, color='blue', alpha=0.2, label='Incerteza')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exemplo de Ãrea Sombreada')
plt.legend()
plt.tight_layout()
plt.show()
````

## **ðŸ“‰GrÃ¡fico com barras de erro - `plt.errorbar()`**

## **ðŸ”¥Imagem ou mapa de calor - `plt.imshow()`**
* Ideal para matrizes, campos, resultados de simulaÃ§Ãµes numÃ©ricas (ex: CFD, transferÃªncia de calor).
* Muito usado para visualizar mapas 2D (temperatura, concentraÃ§Ã£o).


````python
# Grade espacial e temporal
x = np.linspace(0, 1, 20)          # posiÃ§Ã£o espacial
t = np.linspace(0, 10, 50)         # tempo
T, X = np.meshgrid(t,x)           # grade para simulaÃ§Ã£o

# Temperatura simulada
temperatura = np.sin(np.pi * X) * np.exp(-0.3 * T)

# VisualizaÃ§Ã£o com imshow
plt.figure(figsize=(8, 5))
im = plt.imshow(temperatura, aspect='auto', extent=[x.min(), x.max(), t.min(), t.max()],
                cmap='inferno')

plt.colorbar(label='Temperatura')
plt.ylabel('PosiÃ§Ã£o (x)')
plt.xlabel('Tempo (t)')
plt.title('Mapa de Calor: Temperatura T(t, x)')
plt.tight_layout()
plt.show()
````

###**ðŸ” CÃ³digo da animaÃ§Ã£o**

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Grade temporal e espacial
x = np.linspace(0, 1, 50)
t = np.linspace(0, 10, 100)
T, X = np.meshgrid(t, x)

# Temperatura simulada
temperatura = np.sin(np.pi * X) * np.exp(-0.3 * T)

# InicializaÃ§Ã£o com um pequeno intervalo de tempo visÃ­vel
delta_t = (t[1] - t[0]) if len(t) > 1 else 0.1

fig, ax = plt.subplots(figsize=(8, 4))
img = ax.imshow(temperatura[:, 0:2], aspect='auto',
                extent=[t[0], t[0] + delta_t, x.min(), x.max()],
                cmap='inferno', vmin=0, vmax=1)

# ConfiguraÃ§Ãµes visuais
ax.set_xlabel('Tempo (t)')
ax.set_ylabel('PosiÃ§Ã£o (x)')
ax.set_title('EvoluÃ§Ã£o da Temperatura T(t, x)')
cbar = fig.colorbar(img, ax=ax)
cbar.set_label('Temperatura')

# FunÃ§Ã£o de atualizaÃ§Ã£o
def update(frame):
    if frame < 2:
        width = delta_t
    else:
        width = t[frame] - t[0]
    img.set_data(temperatura[:, 0:frame+1])
    img.set_extent([t[0], t[0] + width, x.min(), x.max()])
    return [img]

# Criar animaÃ§Ã£o
ani = animation.FuncAnimation(fig, update, frames=len(t), interval=100, blit=True)

plt.close(fig)  # evita mostrar a figura duplicada

# Exibir no notebook
HTML(ani.to_jshtml())
````

#### âœ… 2. GrÃ¡fico de dispersÃ£o com `sns.scatterplot()`

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criar DataFrame com grupos
np.random.seed(42)
n = 50
df = pd.DataFrame({
    'vazao': np.random.normal(100, 10, n),
    'concentracao': np.random.normal(0.5, 0.1, n),
    'grupo': np.random.choice(['A', 'B'], n)
})

# GrÃ¡fico de dispersÃ£o
sns.scatterplot(data=df, x='vazao', y='concentracao', hue='grupo', style='grupo')
plt.title('RelaÃ§Ã£o VazÃ£o x ConcentraÃ§Ã£o')
plt.xlabel('VazÃ£o (kg/h)')
plt.ylabel('ConcentraÃ§Ã£o (fraÃ§Ã£o mÃ¡ssica)')
plt.grid(True)
plt.tight_layout()
plt.show()
````

#### âœ… 3. GrÃ¡fico de mapa de calor com `sns.heatmap()`

* Ã© uma ferramenta utilizada para visualizar valores organizados em uma matriz, permitindo identificar rapidamente padrÃµes, correlaÃ§Ãµes e regiÃµes crÃ­ticas.


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com dados simulados
dados = {
    8:  [1.249080, 0.812037, 1.702230],
    12: [2.401429, 0.811989, 1.916145],
    16: [1.963988, 0.616167, 40.000000],
    20: [1.697317, 2.232352, 2.439820],
}

indice = [3, 4, 5]
df = pd.DataFrame(dados, index=indice)
df.index.name = "NÂº de neurÃ´nios"
df.columns.name = "Camadas ocultas"

# Plotando o heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df, annot=True, fmt=".2f", cmap="coolwarm", vmax=5, cbar_kws={'label': 'Erro QuadrÃ¡tico MÃ©dio'})
plt.title("Erro quadrÃ¡tico mÃ©dio dos modelos (valores simulados)")
plt.xlabel("Camadas ocultas")
plt.ylabel("NÂº de neurÃ´nios")
plt.tight_layout()
plt.show()
````

````python
# Passo 2: Exemplo de GrÃ¡fico Interativo
import plotly.graph_objects as go
import numpy as np

# Dados de exemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Criar figura com duas curvas
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Seno',
                         line=dict(color='blue')))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Cosseno',
                         line=dict(color='red', dash='dash')))

# PersonalizaÃ§Ãµes
fig.update_layout(
    title='GrÃ¡fico Interativo com Plotly',
    xaxis_title='x',
    yaxis_title='FunÃ§Ã£o',
    legend_title='FunÃ§Ãµes',
    template='plotly_dark',  # ou plotly_white, ggplot2, etc.
    hovermode='x unified'
)

# Exibir no Colab
fig.show()
````

````python
# Passo 2: Exemplo de GrÃ¡fico Interativo

import numpy as np
import pyvista as pv
from IPython.display import Image

# Gerar dados
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# SuperfÃ­cie
z = np.sin(3 * np.sqrt(x**2 + y**2))/ (1 + x**2 + y**2)

# Criar grade estruturada
grid = pv.StructuredGrid(x, y, np.zeros_like(z))  # Altura inicial plana
grid["altura"] = z.ravel(order="F")  # Adiciona como escalar

# Deformar a superfÃ­cie com base no escalar
surface = grid.warp_by_scalar("altura", factor=1.0)

# Criar plotter off-screen
plotter = pv.Plotter(off_screen=True)
plotter.add_mesh(surface, cmap='inferno', show_edges=False)
plotter.add_axes()
plotter.add_text("SuperfÃ­cie: z = f(x, y)", font_size=12)
plotter.set_background("white")
plotter.show(screenshot='superficie.png')

# Exibir imagem
Image('superficie.png')
````
