# 🎨 Introdução à Visualização de Dados com Matplotlib

A biblioteca **Matplotlib** é uma das ferramentas mais populares em Python para criação de gráficos. Ela permite representar dados de forma visual, o que facilita a análise, a comunicação e a compreensão de informações complexas.

## 🧭 Por que usar gráficos?

- 📈 Compreender **tendências e padrões**
- 🔍 Investigar **relações entre variáveis**
- 🧪 Visualizar **resultados de experimentos ou simulações**
- 🧠 Apoiar **decisões técnicas** e **comunicação científica**

## 🧩 Anatomia de um gráfico

Um gráfico em Matplotlib geralmente envolve:

- **Figura (`Figure`)**: a área total de desenho (janela do gráfico)
- **Eixos (`Axis`)**: região onde os dados são desenhados (com eixos x e y)
- **Curvas e marcadores (`Lines`)**: os dados propriamente ditos
- **Rótulos (`Labels`)**: nomes dos eixos, título e legendas
- **Grade (`Grid`)**: linhas de referência opcionais
- **Barra de cores (`Colorbar`)**: útil em mapas de calor

## 🛠️ Fluxo básico para criar um gráfico

# **📈 Gráfico de linhas - `plt.plot()`**

O `plt.plot()` é usado para visualizar dados contínuos, como variações ao longo do tempo ou resultados de simulações.

Este exemplo mostra como sobrepor duas curvas simples, aplicando boas práticas de visualização: uso de `label`, `grid`, título, rótulos de eixos e `tight_layout()` para melhor espaçamento.



````python
# importar as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# criar variável x (independente)
var_x = np.array([0, 1, 2, 3])

# criar variáveis y (dependentes)
y1 = [0, 1, 2, 3]                      # função linear: y = x
y2 = [x**2 + 1 for x in var_x]         # função quadrática: y = x² + 1

# criar a figura (tamanho opcional)
plt.figure(1, figsize=(6, 3))

# plotar os gráficos com rótulos e estilos
plt.plot(var_x, y1, 'b-', label='Linear')        # linha azul contínua
plt.plot(var_x, y2, 'r--', label='Quadrática')   # linha vermelha tracejada

# customização dos eixos e título
plt.xlabel('Variável independente')
plt.ylabel('Valor de resposta')
plt.title('Exemplo de Gráfico')

# exibir legenda e grade
plt.legend()
plt.grid(True)

# ajustar o layout e exibir
plt.tight_layout()
plt.show()
````

# **Múltiplos gráficos - `plt.subplot(m,n,p)`**

O comando `plt.subplot(m, n, p)` divide a figura em uma grade de `m` linhas e `n` colunas, e ativa a posição `p` para desenhar o gráfico. Isso é útil para comparar diferentes visualizações lado a lado.

````python
# Criar vetor de entrada
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)          # função seno
y2 = np.cos(x)          # função cosseno
y3 = np.sin(x) + np.cos(x)  # combinação das duas

# Criar figura
plt.figure(figsize=(6, 6))

# Gráfico 1: seno
plt.subplot(3, 1, 1)  # 3 linhas, 1 coluna, 1ª posição
plt.plot(x, y1, 'b')
plt.title('Função seno')
plt.ylabel('sin(x)')

# Gráfico 2: cosseno
plt.subplot(3, 1, 2)  # 3 linhas, 1 coluna, 2ª posição
plt.plot(x, y2, 'r')
plt.title('Função cosseno')
plt.ylabel('cos(x)')

# Gráfico 3: soma
plt.subplot(3, 1, 3)  # 3 linhas, 1 coluna, 3ª posição
plt.plot(x, y3, 'g')
plt.title('Seno + Cosseno')
plt.xlabel('x')
plt.ylabel('sin(x) + cos(x)')

# Ajuste automático para evitar sobreposição
plt.tight_layout()
plt.show()
````

# **Gráfico em 2D/3D**

## **Como usar `plt.contourf()` e `plot_surface`?**

> 💡 Explicação Rápida:
* `contourf` → visualização 2D com níveis de cor preenchidos.
* `plot_surface` → representação 3D da mesma função.
* `cm.viridis_r` → colormap invertido para melhor contraste.

**Exemplo simples**

````python
Z = np.random.uniform(0,1,(8,8))
plt.contourf(Z, cmap='viridis_r')
plt.colorbar()
plt.show()
````

**Exemplo com função $Z=f(X,Y)$**

````python

from mpl_toolkits.mplot3d import Axes3D

# Criar a grade de pontos (x, y)
nx, ny = 100, 100
x = np.linspace(0, 0.909, nx)
y = np.linspace(0, 0.909, ny)
X, Y = np.meshgrid(x, y)

# Função paraboloide deslocado
Z = (X - 0.2)**2 + (Y - 0.2)**2

# Criar figura com 2 subplots lado a lado
fig = plt.figure(figsize=(12, 5))

# Subplot 1: contorno (2D)
ax1 = fig.add_subplot(1, 2, 1)
cont = ax1.contourf(x, y, Z, cmap='viridis_r')  # mapa de cores
cbar1 = fig.colorbar(cont, ax=ax1)  # barra de cores
cbar1.set_label('z = f(x, y)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Mapa de Contorno')

# Subplot 2: superfície (3D)
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
surf = ax2.plot_surface(X, Y, Z, cmap='viridis_r')
cbar2 = fig.colorbar(surf, ax=ax2, shrink=0.5, aspect=10, pad=0.1)
cbar2.set_label('z = f(x, y)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z = f(x, y)')
ax2.set_title('Paraboloide')

# Ajuste de layout para evitar sobreposição
plt.tight_layout()
plt.show()
````

# **Outros tipos de gráficos**
## **🔸 Gráfico de Dispersão - `plt.scatter()`**

O `plt.scatter()` é útil para visualizar a relação entre duas variáveis, especialmente em dados experimentais com ruído.

É ideal para detectar correlações, padrões ou agrupamentos nos dados.

````python
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)  # y com ruído

plt.scatter(x, y, color='darkorange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exemplo de Gráfico de Dispersão')
plt.grid(True)
plt.tight_layout()
plt.show()
````

## **📊 Histograma- `plt.hist()`**

O `plt.hist()` mostra a distribuição de frequências de um conjunto de dados, agrupando os valores em "bins".

É uma ferramenta importante para análises estatísticas iniciais, como avaliação de normalidade.

````python
dados = np.random.normal(0, 1, 1000)

plt.hist(dados, bins=20, color='steelblue', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Exemplo de Histograma')
plt.tight_layout()
plt.show()
````

## **📦 Boxplot - `plt.boxplot()`**

O `plt.boxplot()` é usado para representar estatísticas descritivas de um conjunto de dados:
- Mediana
- Quartis
- Outliers (valores extremos)

É muito útil para comparar a variabilidade entre diferentes grupos experimentais.

````python
grupos = [np.random.normal(loc, 1, 100) for loc in [0, 1, 2]]

plt.boxplot(grupos, tick_labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.ylabel('Valor')
plt.title('Exemplo de Boxplot')
plt.tight_layout()
plt.show()
````

## **📊 Gráfico de barras - `plt.bar()`**

O `plt.bar()` é usado para visualizar comparações entre categorias.  
É útil para mostrar contagens, médias ou totais agrupados.

Neste exemplo, cada letra representa uma categoria e a altura da barra representa o valor correspondente.

````python
# categorias e valores associados
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 12, 31]

# criar o gráfico de barras
plt.bar(categorias, valores, color='skyblue')  # barras azuis claras

# adicionar rótulos e título
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Exemplo de Gráfico de Barras')

# exibir o gráfico
plt.tight_layout()
plt.show()
````

## **🍕 Gráfico de Pizza - `plt.pie()`**

O `plt.pie()` representa proporções relativas entre categorias.  
É mais eficaz quando há poucas categorias (tipicamente até 5 ou 6).

Cada "fatia" mostra a participação percentual de cada grupo no total.

````python
valores = [40, 25, 20, 15]
categorias = ['A', 'B', 'C', 'D']

plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90)
plt.title('Exemplo de Gráfico de Pizza')
plt.tight_layout()
plt.show()
````

## **🟦 Área Sombreada - `plt.fill_between()`**


O `plt.fill_between()` preenche a área entre duas curvas ou entre uma curva e um valor fixo.

É ideal para mostrar:
- Incertezas experimentais
- Intervalos de confiança
- Faixas de operação segura

````python
x = np.linspace(0, 10, 100)
y_central = np.sin(x)
y_sup = y_central + 0.3
y_inf = y_central - 0.3

plt.plot(x, y_central, label='Média', color='blue')
plt.fill_between(x, y_inf, y_sup, color='blue', alpha=0.2, label='Incerteza')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exemplo de Área Sombreada')
plt.legend()
plt.tight_layout()
plt.show()
````

## **📉Gráfico com barras de erro - `plt.errorbar()`**
* Útil para mostrar medidas experimentais com incerteza.
* Exemplo clássico em engenharia ao lidar com médias e desvio padrão.


````python
x = np.array([1, 2, 3, 4])
y = np.array([10, 15, 13, 17])
erro = np.array([1.5, 1.0, 0.8, 1.2])

plt.errorbar(x, y, yerr=erro, fmt='o', capsize=5)
plt.title("Gráfico com Barras de Erro")
plt.xlabel("Tempo")
plt.ylabel("Valor medido")
plt.grid(True)
plt.tight_layout()
plt.show()
````

## **🔥Imagem ou mapa de calor - `plt.imshow()`**
* Ideal para matrizes, campos, resultados de simulações numéricas (ex: CFD, transferência de calor).
* Muito usado para visualizar mapas 2D (temperatura, concentração).


````python
# Grade espacial e temporal
x = np.linspace(0, 1, 20)          # posição espacial
t = np.linspace(0, 10, 50)         # tempo
T, X = np.meshgrid(t,x)           # grade para simulação

# Temperatura simulada
temperatura = np.sin(np.pi * X) * np.exp(-0.3 * T)

# Visualização com imshow
plt.figure(figsize=(8, 5))
im = plt.imshow(temperatura, aspect='auto', extent=[x.min(), x.max(), t.min(), t.max()],
                cmap='inferno')

plt.colorbar(label='Temperatura')
plt.ylabel('Posição (x)')
plt.xlabel('Tempo (t)')
plt.title('Mapa de Calor: Temperatura T(t, x)')
plt.tight_layout()
plt.show()
````

###**🔁 Código da animação**

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Grade temporal e espacial
x = np.linspace(0, 1, 50)
t = np.linspace(0, 10, 100)
T, X = np.meshgrid(t, x)

# Temperatura simulada
temperatura = np.sin(np.pi * X) * np.exp(-0.3 * T)

# Inicialização com um pequeno intervalo de tempo visível
delta_t = (t[1] - t[0]) if len(t) > 1 else 0.1

fig, ax = plt.subplots(figsize=(8, 4))
img = ax.imshow(temperatura[:, 0:2], aspect='auto',
                extent=[t[0], t[0] + delta_t, x.min(), x.max()],
                cmap='inferno', vmin=0, vmax=1)

# Configurações visuais
ax.set_xlabel('Tempo (t)')
ax.set_ylabel('Posição (x)')
ax.set_title('Evolução da Temperatura T(t, x)')
cbar = fig.colorbar(img, ax=ax)
cbar.set_label('Temperatura')

# Função de atualização
def update(frame):
    if frame < 2:
        width = delta_t
    else:
        width = t[frame] - t[0]
    img.set_data(temperatura[:, 0:frame+1])
    img.set_extent([t[0], t[0] + width, x.min(), x.max()])
    return [img]

# Criar animação
ani = animation.FuncAnimation(fig, update, frames=len(t), interval=100, blit=True)

plt.close(fig)  # evita mostrar a figura duplicada

# Exibir no notebook
HTML(ani.to_jshtml())
````

#### ✅ 1. Gráfico de linha com `sns.lineplot()`

* ideal para séries temporais ou simulações

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Criar dados simulados
x = np.linspace(0, 10, 100)
dados = pd.DataFrame({
    'tempo': x,
    'temperatura': np.sin(x) + 0.2*np.random.randn(100)
})

# Gráfico de linha com Seaborn
sns.lineplot(data=dados, x='tempo', y='temperatura', color='crimson')
plt.title('Variação da Temperatura no Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
````

#### ✅ 2. Gráfico de dispersão com `sns.scatterplot()`

* ideal para análise de correlação e agrupamentos



````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criar DataFrame com grupos
np.random.seed(42)
n = 50
df = pd.DataFrame({
    'vazao': np.random.normal(100, 10, n),
    'concentracao': np.random.normal(0.5, 0.1, n),
    'grupo': np.random.choice(['A', 'B'], n)
})

# Gráfico de dispersão
sns.scatterplot(data=df, x='vazao', y='concentracao', hue='grupo', style='grupo')
plt.title('Relação Vazão x Concentração')
plt.xlabel('Vazão (kg/h)')
plt.ylabel('Concentração (fração mássica)')
plt.grid(True)
plt.tight_layout()
plt.show()
````

#### ✅ 3. Gráfico de mapa de calor com `sns.heatmap()`

* é uma ferramenta utilizada para visualizar valores organizados em uma matriz, permitindo identificar rapidamente padrões, correlações e regiões críticas.


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com dados simulados
dados = {
    8:  [1.249080, 0.812037, 1.702230],
    12: [2.401429, 0.811989, 1.916145],
    16: [1.963988, 0.616167, 40.000000],
    20: [1.697317, 2.232352, 2.439820],
}

indice = [3, 4, 5]
df = pd.DataFrame(dados, index=indice)
df.index.name = "Nº de neurônios"
df.columns.name = "Camadas ocultas"

# Plotando o heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df, annot=True, fmt=".2f", cmap="coolwarm", vmax=5, cbar_kws={'label': 'Erro Quadrático Médio'})
plt.title("Erro quadrático médio dos modelos (valores simulados)")
plt.xlabel("Camadas ocultas")
plt.ylabel("Nº de neurônios")
plt.tight_layout()
plt.show()
````

````python
# Passo 2: Exemplo de Gráfico Interativo

import numpy as np
import pyvista as pv
from IPython.display import Image

# Gerar dados
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# Superfície
z = np.sin(3 * np.sqrt(x**2 + y**2))/ (1 + x**2 + y**2)

# Criar grade estruturada
grid = pv.StructuredGrid(x, y, np.zeros_like(z))  # Altura inicial plana
grid["altura"] = z.ravel(order="F")  # Adiciona como escalar

# Deformar a superfície com base no escalar
surface = grid.warp_by_scalar("altura", factor=1.0)

# Criar plotter off-screen
plotter = pv.Plotter(off_screen=True)
plotter.add_mesh(surface, cmap='inferno', show_edges=False)
plotter.add_axes()
plotter.add_text("Superfície: z = f(x, y)", font_size=12)
plotter.set_background("white")
plotter.show(screenshot='superficie.png')

# Exibir imagem
Image('superficie.png')
````

````python
from pyvista import examples

# Baixar malha do coelho de Stanford
mesh = examples.download_bunny_coarse()

# Criar plotter off-screen
pl = pv.Plotter(off_screen=True)
_ = pl.add_mesh(mesh, show_edges=True, reset_camera=True)

# Definir posição da câmera (vista lateral superior)
pl.camera_position = [
    (0.3914, 0.4542, 0.7670),  # posição da câmera
    (0.0243, 0.0336, -0.0222),  # ponto focal (olhando para)
    (-0.2148, 0.8998, -0.3796),  # vetor para cima
]

# Salvar imagem e exibir no Colab
pl.show(screenshot='bunny.png')
Image('bunny.png')
````
