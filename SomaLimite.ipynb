# Pseudocódigo

Definir soma_limite ← π / 4

Definir n ← 0

Definir soma ← 0

Definir erro ← valor grande (ex: 1000)

Enquanto erro > 10^-6 faça:

* Calcular termo ← (-1)^n / (2*n + 1)

* Atualizar soma ← soma + termo

* Calcular erro ← (soma - soma_limite)^2

* Atualizar n ← n + 1

Fim enquanto

Exibir: "O somatório pode ser aproximado com", n, "termos."

```python
from graphviz import Digraph

dot = Digraph()

dot.attr(rankdir='TB')

dot.node('inicio', 'Início', shape='ellipse')
dot.node('init', 'soma_limite = π/4\nn = 0\nsoma = 0\nerro = grande')
dot.node('decisao', 'erro > 1e-6 ?', shape='diamond')
dot.node('termo', 'termo = (-1)^n / (2n + 1)')
dot.node('soma', 'soma = soma + termo')
dot.node('erro', 'erro = (soma - soma_limite)^2')
dot.node('atualiza', 'n = n + 1')
dot.node('saida', 'Exibir n (nº de termos)')
dot.node('fim', 'Fim', shape='ellipse')

dot.edge('inicio', 'init')
dot.edge('init', 'decisao')

dot.edge('decisao', 'termo', label='Sim')
dot.edge('termo', 'soma')
dot.edge('soma', 'erro')
dot.edge('erro', 'atualiza')
dot.edge('atualiza', 'decisao')

dot.edge('decisao', 'saida', label='Não')
dot.edge('saida', 'fim')

dot
```

# Solução computacional


```python
import numpy as np
import matplotlib.pyplot as plt

soma_limite = np.pi / 4
n = 0  # primeiro valor do contador do somatório
s = []  # lista dos elementos do somatório
soma = 0.0 # inicia a soma como elemento neutro
erro = 1000  # erro inicial (valor arbitrário)
max_iter = 1000 # número máximo de iterações

while erro > 1e-6 and n<max_iter:
    termo = (-1)**n / (2*n + 1)
    s.append(termo)  # acumulando os termos da série
    soma = soma + termo  # fazendo a soma acumulada dos termos
    erro = (soma - soma_limite)**2 # erro quadrático
    n = n + 1
else:
    if n == max_iter:
      print("Não convergiu dentro do limite de iterações.")
    else:
      print('O somatório pode ser aproximado com', n, 'termos.')

plt.plot(np.cumsum(s))
plt.plot(np.ones(n) * soma_limite, '--r')
plt.legend(('Soma aproximada', 'Limite da soma'))
plt.xlabel('Número de termos')
plt.ylabel('Valor acumulado')
plt.title(f'Convergência da série para pi/4 usando {n} termos')
plt.grid()
plt.show()
```
