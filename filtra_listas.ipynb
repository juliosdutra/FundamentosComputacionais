# Definindo a função

O código deve apresentar o seguintes conteúdos:

1. **Anotações de tipo**
2. Estrutura `try-except` para validações.
3. **Mensagens informativas**.
4. `raise ValueError` quando necessário.


```python
def filtra_numeros(lista: list, limite_inferior: int = 0, limite_superior: int = 100, apenas_pares: bool = False) -> list:
    """
    Filtra os números de uma lista que estão dentro de um intervalo e, opcionalmente, apenas os pares.

    Args:
        lista (list): Lista de números a serem filtrados.
        limite_inferior (int): Valor mínimo do intervalo (inclusivo).
        limite_superior (int): Valor máximo do intervalo (inclusivo).
        apenas_pares (bool): Se True, retorna apenas os pares dentro do intervalo.

    Returns:
        list: Lista de números filtrados conforme os critérios definidos.
    """
    try:
        if not isinstance(lista, list):
            raise ValueError("O parâmetro 'lista' deve ser uma lista.")

        resultado = []

        for item in lista:
            if not isinstance(item, (int, float)):
                raise ValueError(f"Item inválido na lista: {item} não é um número.")

            if limite_inferior <= item <= limite_superior:
                if apenas_pares:
                    if item % 2 == 0:
                        resultado.append(item)
                else:
                    resultado.append(item)

        print(f"Lista filtrada com sucesso: {resultado}")
        return resultado

    except Exception as e:
        print(f"Erro ao filtrar a lista. {e}")
        return []
```

# Testes

```python
filtra_numeros([10, 23, 45, 60, 80, 110], limite_inferior=20, limite_superior=90, apenas_pares=False)
```

```python
filtra_numeros([10, 23, 45, 60, 80, 110], limite_inferior=20, limite_superior=90, apenas_pares=True)
```
