```python
def ajustar_temperatura_reacao():
    try:
        temp_ambiente_str = input("Digite a temperatura ambiente (°C): ").strip()
        temp_ambiente = float(temp_ambiente_str)

        if temp_ambiente < 15:
            temp_reacao = temp_ambiente + 10
            acao = "aumentada em 10 °C"
        elif temp_ambiente <= 30:
            temp_reacao = temp_ambiente
            acao = "mantida igual"
        else:
            temp_reacao = temp_ambiente - 5
            acao = "reduzida em 5 °C"

        print(f"A temperatura de reação será {acao}, resultando em {temp_reacao:.1f} °C.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número válido para a temperatura.")

# Chamada da função
ajustar_temperatura_reacao()
```
